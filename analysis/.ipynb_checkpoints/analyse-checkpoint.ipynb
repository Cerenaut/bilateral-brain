```python
import numpy as np
import scipy as sp
from itertools import groupby, combinations
from operator import itemgetter
from sklearn.cluster import KMeans
from sklearn.neighbors import KNeighborsClassifier
```

```python
narrow_feat = np.load('./narrow_feat.npy').reshape(-1, 512)
broad_feat = np.load('./broad_feat.npy').reshape(-1, 512)
bicameral_feat = np.load('./bicameral_feat.npy').reshape(-1, 1024)
print(narrow_feat.shape, broad_feat.shape, bicameral_feat.shape)
```

```python
filetype = 'nfbtbicamt'
with open(f'./distribution_images_{filetype}.txt', 'r') as f:
    images = f.read().splitlines()
images = [(ind, image.split('/')[-2]) for ind, image in enumerate(images)]
```

```python
grouped_images = [list(group) for key, group in groupby(images, itemgetter(1))]
print(grouped_images)
```

```python
narrow_distances = []
broad_distances = []
bicameral_distances = []
ensemble_distances = []
labels = []
for group in grouped_images:
    if len(group) <= 1:
        narrow_distances.append([])
        broad_distances.append([])
        bicameral_distances.append([])
        ensemble_distances.append([])
        continue
    narrow_matrix = np.zeros((len(group), len(group)))
    broad_matrix = np.zeros((len(group), len(group)))
    bicameral_matrix = np.zeros((len(group), len(group)))
    ensemble_matrix = np.zeros((len(group), len(group)))
    labels.append(int(group[0][1]))
    for pair in list(combinations(enumerate(group), 2)):
        narrow_matrix[pair[0][0]][pair[1][0]] = sp.spatial.distance.cosine(
                                                narrow_feat[pair[0][1][0]] , 
                                                 narrow_feat[pair[1][1][0]])
        broad_matrix[pair[0][0]][pair[1][0]] = sp.spatial.distance.cosine(
                                                broad_feat[pair[0][1][0]], 
                                                 broad_feat[pair[1][1][0]])
        bicameral_matrix[pair[0][0]][pair[1][0]] = sp.spatial.distance.cosine(
                                                bicameral_feat[pair[0][1][0]], 
                                                 bicameral_feat[pair[1][1][0]])
        ensemble_matrix[pair[0][0]][pair[1][0]] = sp.spatial.distance.cosine(
                                        np.hstack([narrow_feat[pair[0][1][0]], broad_feat[pair[0][1][0]]]), 
                                        np.hstack([narrow_feat[pair[1][1][0]], broad_feat[pair[1][1][0]]]))
    narrow_distances.append(narrow_matrix)
    broad_distances.append(broad_matrix)
    bicameral_distances.append(bicameral_matrix)
    ensemble_distances.append(ensemble_matrix)
```

```python
ind = 0

for i in range(len(narrow_distances)):
    if len(narrow_distances[i]) == 0:
        continue
    print(f"Label - {labels[ind]}")
    ind += 1
    print()
    print("i <-> j -- Narrow Distance    =   Broad Distance   =   Bicameral Distance  =  Combining Broad and Narrow w/o Bicameral")
    for j in range(narrow_distances[i].shape[0]):
        for k in range(narrow_distances[i].shape[1]):
            if narrow_distances[i][j][k] == 0:
                continue
            print(f"{j} <-> {k} -- {narrow_distances[i][j][k]} = {broad_distances[i][j][k]} = {bicameral_distances[i][j][k]} = {ensemble_distances[i][j][k]}")
    print("-------------------------\n\n")
```
